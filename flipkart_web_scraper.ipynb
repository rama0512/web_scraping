```python
#fetch links as list of tag object 
links=soup.find_all("div", attrs={'class':'_4ddWXP'})
```

```python
df=pd.DataFrame(columns=['name','link','price','rating','rating_volumee'])
df
```

```python
for link in links:
    name_data=link.find("a",attrs={"class":"s1Q9rs"}).get("title")
    link_data='https://www.flipkart.com'+link.find("a",attrs={"class":"s1Q9rs"}).get("href")
    price_data=link.find("div",attrs={"class":"_30jeq3"}).text
    rating_data=link.find("div",attrs={"class":"_3LWZlK"})
    rating_volume=link.find("span",attrs={"class":"_2_R_DZ"})
    row_data=[name_data,link_data,price_data,rating_data,rating_volume]
    length=len(df)
    df.loc[length] = row_data  
```
